```python
import pandas as pd
import numpy as np
import spikeinterface as si
import spikeinterface.extractors as se
import spikeinterface.preprocessing as spre
from pathlib import Path
from utils import *
import os
import phy.apps.kwik
import phy.apps.template

%matplotlib widget


sheet = pd.read_excel('probe_data_IV.xlsx')
sheet_inc = sheet[sheet['include_spikeinterface'] == 'Y']
sorting_suffix = 'sorting_klusta_4_7'


trial_list = sheet_inc['trial_name']. to_list()
session_list = np.unique([i[:6] for i in trial_list])    

# Collect into sessions and preprocess
recording_list = []
for i,session in enumerate(session_list):
    recording_list.append([])
    for j in trial_list:
        if session in j:
            base_folder = sheet_inc[sheet_inc['trial_name'] == j]['path'].tolist()[0]
            num_channels = int(sheet_inc[sheet_inc['trial_name'] == j]['num_channels'].tolist()[0])
            electrode_type = sheet_inc[sheet_inc['trial_name'] == j]['probe_type'].tolist()[0]
            print(f"{base_folder}/{j}.set")
            recording = se.read_axona(f"{base_folder}/{j}.set")
            
            #Invert trace if recorded on -signal or ref-signal mode (mode 2 and 3)
#             if int(get_mode(f"{base_folder}/{j}.set")) in [2,3]:
#                 recording = spre.scale(recording, gain=-1)
#                 print('Inverting recording trace')
            
            
            recording_list[i].append([preprocess(recording = recording,
                                     recording_name = j,
                                     base_folder = base_folder,
                                     electrode_type = electrode_type,
                                     num_channels = num_channels),
                                     j, base_folder, electrode_type])
                                     
                
# Concatenate over a single session and sort
recordings_concat = []
for i, recording in enumerate(recording_list):
    session = pd.DataFrame(recording_list[i])
    
    recordings_concat.append(si.concatenate_recordings(session.iloc[:,0]))
    for j, concat in enumerate(recordings_concat):
        print(f'Sorting {concat}')
        sorting = sort(recording = concat, 
             recording_name = session_list[j], 
             base_folder = session.iloc[0,2],
             electrode_type = session.iloc[0,3],
             sorting_suffix = sorting_suffix)
        session.to_csv(f'{base_folder}/{session_list[j]}_{sorting_suffix}/session.csv') #save session trial info to .csv
        
#         electrode_type = session.iloc[0,3]
        
#         if electrode_type == 'tetrode':
#             recording_to_process = concat
#             recording_f = spre.bandpass_filter(recording_to_process, freq_min=300, freq_max=6000)
#             recording_cmr = spre.common_reference(recording_f, reference='global', operator='median')
            
#             #Save spikeinterface-extracted waveforms
#             print(sorting)
#             we = si.extract_waveforms(recording_cmr, sorting,
#                                       folder= '/home/isabella/Documents/isabella/jake/recording_data/waveforms', 
#                                       load_if_exists=False, overwrite=True)
#             print(we)

#             # Export to phy
#             sexp.export_to_phy(we, output_folder= f'{sorting_path}/phy', 
#                       compute_amplitudes=True, compute_pc_features=True, copy_binary=True, peak_sign = 'neg', verbose = True)

print(session)
```
